```python
import requests
from bs4 import BeautifulSoup

# Define the sitemap URL
# sitemap_url = 'https://www.davidkolbconsultancy.com/pages-sitemap.xml'

sitemap_url = 'https://www.davidkolbconsultancy.com/blog-posts-sitemap.xml'

# Fetch the sitemap content
response = requests.get(sitemap_url)
sitemap_content = response.content

# Parse the sitemap using BeautifulSoup
soup = BeautifulSoup(sitemap_content, 'xml')

# Extract all URLs from the sitemap
urls = [loc.text for loc in soup.find_all('loc')]

# For each URL, fetch content and extract h1, h2, h3, meta title, and meta description
count = 0 
for url in urls:
    count += 1
    response = requests.get(url)
    page_content = response.content
    page_soup = BeautifulSoup(page_content, 'html.parser')
    
    # Meta title and description
    meta_title = page_soup.title.string if page_soup.title else None
    meta_description_tag = page_soup.find('meta', attrs={'name': 'description'})
    meta_description = meta_description_tag['content'] if meta_description_tag else None
    
    if meta_title:
        print(f"URL: {url} - Meta Title: {meta_title}")
    if meta_description:
        print(f"URL: {url} - Meta Description: {meta_description}")

    if count == 10:
        print("Counter reached 10")
        count = 0  # Reset the counter to 0
        
        
    # Extract h1, h2, and h3 tags
    #   h1_tags = page_soup.find_all('h1')
 #      for h1 in h1_tags:
     #      print(f"URL: {url} - H1: {h1.text}")

    #   h2_tags = page_soup.find_all('h2')
    #   for h2 in h2_tags:
     #      print(f"URL: {url} - H2: {h2.text}")

   #    h3_tags = page_soup.find_all('h3')
   #    for h3 in h3_tags:
    #       print(f"URL: {url} - H3: {h3.text}")
```
