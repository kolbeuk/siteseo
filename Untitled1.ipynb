```python
# Function to count characters in meta titles and descriptions
def count_characters(file_path):
    # Initialize variables to store counts
    title_count = 0
    description_count = 0

    try:
        # Open the file for reading
        with open(file_path, 'r') as file:
            lines = file.readlines()

            # Initialize variables to store current meta title and description
            current_title = ""
            current_description = ""

            # Iterate through the lines in the file
            for line in lines:
                # Check if the line contains "Meta Title: "
                if "Meta Title: " in line:
                    # Print the current meta title and character count
                    if current_title:
                        print(f"Meta Title: {current_title}")
                        print(f"Character Count: {title_count}")
                    # Extract the new meta title text
                    current_title = line.replace("Meta Title: ", "")
                    # Reset character count
                    title_count = 0

                # Check if the line contains "Meta Description: "
                if "Meta Description: " in line:
                    # Print the current meta description and character count
                    if current_description:
                        print(f"Meta Description: {current_description}")
                        print(f"Character Count: {description_count}")
                    # Extract the new meta description text
                    current_description = line.replace("Meta Description: ", "")
                    # Reset character count
                    description_count = 0

                # Count characters in the current meta title and description
                if current_title:
                    title_count += len(line.strip())
                if current_description:
                    description_count += len(line.strip())

            # Print the last meta title and description
            if current_title:
                print(f"Meta Title: {current_title}")
                print(f"Character Count: {title_count}")
            if current_description:
                print(f"Meta Description: {current_description}")
                print(f"Character Count: {description_count}")

    except FileNotFoundError:
        print(f"File '{file_path}' not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Specify the path to your text file
file_path = "input.txt"  # Update with your file path

# Call the function to count characters
count_characters(file_path)
```
